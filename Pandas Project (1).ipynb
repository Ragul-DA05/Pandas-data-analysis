```python
#“This project aims to analyze historical stock market data of hdfcbank to understand trading behavior, price trends, and market activity over time.
import pandas as pd
```

```python
#Dropping unwanted column
df.drop(columns="Unnamed: 0",inplace=True)
```

```python
df.head()
```

```python
#converting date to datetime datatype
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
#setting date as my index for easy operation
df.set_index("Date", inplace=True)
```

```python
#sorting in order 
df.sort_index(inplace=True)
df
```

```python
#checking nulls 
df.isnull().sum()
```

```python
#handling null using fill because its trade series
df["Trades"]=df["Trades"].ffill()
df
```

```python
df["Trades"]=df["Trades"].bfill()
df
```

```python
df['MA_20'] = df['MA_20'].fillna(0)
df['MA_50'] = df['MA_50'].fillna(0)
```

```python
df['PrevClose'] = df['PrevClose'].fillna(0)
df['Daily_Return_%'] = df['Daily_Return_%'].fillna(0)
```

```python
df.isnull().sum()
```

```python
#checking outliers using boxplot
import seaborn as sns

sns.boxplot(x="Volume",data=df)
```

```python
sns.boxplot(x="Trades",data=df)
```

```python
import matplotlib.pyplot as plt

# Yearly averages
yearly_trades = df.groupby(df.index.year)['Trades'].mean()
yearly_volume = df.groupby(df.index.year)['Volume'].mean()

# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# ---- Plot 1: Average Trades ----
axes[0].bar(
    yearly_trades.index,
    yearly_trades.values,
    color='#F4A896',       # coral–peach
    edgecolor='#C96B5A',
    linewidth=1
)
axes[0].set_title('Year-wise Average Number of Trades', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Average Trades', fontsize=12)
axes[0].grid(axis='y', linestyle='--', alpha=0.4)

# ---- Plot 2: Average Volume ----
axes[1].bar(
    yearly_volume.index,
    yearly_volume.values,
    color='#A8D5BA',       # pista green
    edgecolor='#4F6F52',
    linewidth=1
)
axes[1].set_title('Year-wise Average Trading Volume', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Year', fontsize=12)
axes[1].set_ylabel('Average Volume', fontsize=12)
axes[1].grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()


```

```python
yearly_high = df.groupby(df.index.year)['High'].max()
yearly_low = df.groupby(df.index.year)['Low'].min()

plt.figure(figsize=(8,5))
plt.plot(yearly_high.index, yearly_high.values, label='Yearly High', color='orange')
plt.plot(yearly_low.index, yearly_low.values, label='Yearly Low', color='green')
plt.title('Year-wise High vs Low Price')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['Month'] = df.index.month
monthly_return = df.groupby('Month')['Daily_Return_%'].mean()

plt.figure(figsize=(8,4))
plt.bar(monthly_return.index, monthly_return.values, color='teal')
plt.title('Average Daily Return by Month')
plt.xlabel('Month')
plt.ylabel('Average Daily Return (%)')
plt.show()
```

```python
df.index = pd.to_datetime(df.index)

df['Decade'] = (df.index.year // 10) * 10
decade_avg_close = df.groupby('Decade')['Close'].mean()

plt.figure(figsize=(8,5))
plt.bar(
    decade_avg_close.index.astype(str),
    decade_avg_close.values,
    color='#A8D5BA',
    edgecolor='#4F6F52',
    linewidth=1
)

plt.title('Decade-wise Average Closing Price', fontsize=14, fontweight='bold')
plt.xlabel('Decade', fontsize=12)
plt.ylabel('Average Close Price', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```

```python
ma_data = df[['MA_20','MA_50']].replace(0, pd.NA)

yearly_ma = ma_data.groupby(df.index.year).mean()

plt.figure(figsize=(8,5))
plt.plot(yearly_ma.index, yearly_ma['MA_20'], label='MA_20')
plt.plot(yearly_ma.index, yearly_ma['MA_50'], label='MA_50')
plt.title('Year-wise Moving Average Trend')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.show()
# by using this we can understand trend direction, whether the market is up or down
```

```python
df['Market Direction'] = df.apply(
    lambda row: 'Bullish Day' if row['Close'] > row['PrevClose'] else 'Bearish Day',
    axis=1
)

direction_counts = df['Market Direction'].value_counts()
plt.figure(figsize=(6,6))
colors = ['#9F86C0', '#CDB4DB']


plt.pie(
    direction_counts,
    labels=direction_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'width': 0.4}
)

plt.title('Bullish vs Bearish Trading Days', fontsize=14, fontweight='bold')
plt.show()
```
